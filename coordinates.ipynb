```python
import pyproj
x_coor = []
y_coor = []
p = pyproj.Proj("+proj=stere +lat_0=90 +lon_0=0.0 +lat_ts=60.0 +a=6378.137 +b=6356.752 +x_0=0 +y_0=0")
for index, row in station_coord.iterrows():
    # print(row['lat'], row['lon'])
    x, y = p(row['lon'], row['lat'], inverse=False)
    
    x2 = round(x * 2) / 2
    x_coor.append(x2)
    
    y2 = round(y * 2) / 2
    y_coor.append(y2)
```

```python
station_coord['x_coor'] = x_coor
```

```python
station_coord['y_coor'] = y_coor
```

```python
station_coord.to_csv('AWS_xy1.csv')
```
